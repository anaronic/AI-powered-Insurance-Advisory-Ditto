REVENUE FORECASTING

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


y = np.array([2020, 2021, 2022, 2023, 2024, 2025]).reshape(-1, 1)  # Reshaping for sklearn
r = np.array([7.66, 9.355, 13.72, 15.84, 21.63, 26.16])  # Revenue data


degree = 3
poly = PolynomialFeatures(degree=degree)
y_poly = poly.fit_transform(y)

# Train Model
model = LinearRegression()
model.fit(y_poly, r)

# Predictions for 2026-2030
future_years = np.array([2026, 2027, 2028, 2029, 2030]).reshape(-1, 1)
future_years_poly = poly.transform(future_years)
predicted_revenue = model.predict(future_years_poly)


for year, rev in zip(range(2026, 2031), predicted_revenue):
    print(f"Predicted Revenue for {year}: {rev:.2f}")


plt.scatter(y, r, color='blue', label="Actual Data")
plt.plot(y, model.predict(y_poly), color='green', linestyle="dashed", label="Fitted Curve")
plt.scatter(future_years, predicted_revenue, color='red', label="Predicted Values")
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.legend()
plt.title("Revenue Prediction for 2026-2030")
plt.show()
```

PROFIT FORECASTING

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Given Data
y = np.array([2020, 2021, 2022, 2023, 2024, 2025]).reshape(-1, 1)
p = np.array([0.0875, -0.4875, -0.545, -1.2, 1.11, 1.43])

# Polynomial Regression
degree = 3
poly = PolynomialFeatures(degree=degree)
y_poly = poly.fit_transform(y)

# Train Model
model = LinearRegression()
model.fit(y_poly, p)

# Predict Profit for 2026-2030
future_years = np.array([2026, 2027, 2028, 2029, 2030]).reshape(-1, 1)
future_years_poly = poly.transform(future_years)
predicted_profit = model.predict(future_years_poly)

for year, profit in zip(range(2026, 2031), predicted_profit):
    print(f"Predicted Profit for {year}: {profit:.2f}")


# Plotting with light blue and sky blue shades
plt.figure(figsize=(8, 5))
plt.scatter(y, p, color='#ADD8E6', label="Actual Data", edgecolors='black', s=100)  # Light blue scatter points
plt.plot(y, model.predict(y_poly), color='#87CEEB', linestyle="dashed", linewidth=2, label="Fitted Curve")  # Sky blue curve
plt.scatter(future_years, predicted_profit, color='#4682B4', label="Predicted Values", edgecolors='black', s=100)  # Steel blue for future points
plt.xlabel("Year")
plt.ylabel("Profit")
plt.legend()
plt.title("Profit Prediction (Light Blue & Sky Blue)")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
